```python
import nltk
nltk.download('punkt')
```

## Let's create a dict of words

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
```

```python
dict_of_words = dict()
ps = PorterStemmer()
sw=set(stopwords.words('spanish'))



for text in df.parent_full_text.values:
    text = str(text)
    for i in [';', ',', ':', ';', '@', '#']:
        text = text.replace(i, '')
        
    text = text.lower()    
    word_tk = word_tokenize(text)
    for word in word_tk:
        if word not in sw:
            
            if ps.stem(word) in dict_of_words.keys():
                dict_of_words[ps.stem(word)] = dict_of_words[ps.stem(word)] + 1

            else:
                dict_of_words[ps.stem(word)] = 1
        
print('There are {} words in this dictionary.'.format(len(dict_of_words)))
```

```python
import json

with open('dict_of_words.json', 'w') as dow:
    json_str = json.dumps(dict_of_words, dow)   
    dow.write(json_str)
```

```python
with open('dict_of_words.json') as json_data:
    dict_of_words = json.load(json_data)
```
