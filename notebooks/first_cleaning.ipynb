### Select original tweets

```python
from copy import deepcopy
original_tweets = deepcopy(tweets[tweets['tweet_type'] == 'original'])
print('There are {} observations (original tweets) in this dataframe with {} variables.'.format(original_tweets.shape[0], tweets.shape[1]))
```

```python
import preprocessor as p
import unidecode

original_tweets['full_text_clean'] = original_tweets['full_text'].apply(lambda x: unidecode.unidecode(p.clean(str(x).lower())))
```

```python
original_tweets.sample()
```

```python
original_tweets['emojis'] = original_tweets['full_text'].apply(lambda x: p.parse(x).emojis)
original_tweets['urls'] = original_tweets['full_text'].apply(lambda x: p.parse(x).urls)
original_tweets['numbers'] = original_tweets['full_text'].apply(lambda x: p.parse(x).numbers) 
original_tweets['hashtags'] = original_tweets['full_text'].apply(lambda x: p.parse(x).hashtags)
```

```python
original_tweets['lo_cuento_yo_porque'] = original_tweets['full_text_clean'].apply(lambda x: 1 if ('lo cuento yo porque' and 'no puede') in str(x).lower() else 0)
```

```python
original_tweets['lamanada'] = original_tweets['full_text_clean'].apply(lambda x: 1 if 'manada' in str(x).lower() else 0)
```

```python
original_tweets['juicio'] = original_tweets['full_text_clean'].apply(lambda x: 1 if 'juicio' in str(x).lower() else 0)
```

```python
original_tweets['ex_pareja'] = original_tweets['full_text_clean'].apply(lambda x: 1 if 'ex pareja' in str(x).lower() else 0)
```

```python
def get_summary_of_tweet(row):
    '''
    '''
    print('Tweet ID: {}'.format(row.id))
    print('Tweet location: {}'.format(row.location))
    print('Tweet created at: {}'.format(row.created_at))
    print('Type of tweet: {}'.format(row.tweet_type))
    print('Tweet: \n {}'.format(row.full_text))
    print('Tweet cleaned: \n {}'.format(row.full_text_clean))
    print('Emojis: {}'.format(row.emojis))
    print('Hashtags: {}'.format(row.hashtags))
    print('URLs: {}'.format(row.urls))
    print('Numbers: {}'.format(row.numbers))
```

```python
get_summary_of_tweet(original_tweets.iloc[1001])
```

## Tweet types

```python
for i in range(50):
    print('-' * 60)
    get_summary_of_tweet(original_tweets.iloc[i])
    print('-' * 60, '\n')
```

```python
print('Tweets with lo cuento yo porque {}'.format(original_tweets['lo_cuento_yo_porque'].sum()))
lo_cuento_yo_tweets = original_tweets[original_tweets['lo_cuento_yo_porque'] == 1]
for i in range(50):
    print('-' * 60)
    get_summary_of_tweet(lo_cuento_yo_tweets.iloc[i])
    print('-' * 60, '\n')
```

```python
print('Tweets with manada {}'.format(original_tweets['lamanada'].sum()))
lamanada_tweets = original_tweets[original_tweets['lamanada'] == 1]
for i in range(50):
    print('-' * 60)
    get_summary_of_tweet(lamanada_tweets.iloc[i])
    print('-' * 60, '\n')
```

```python
print('Tweets with juicio {}'.format(original_tweets['juicio'].sum()))
juicio_tweets = original_tweets[original_tweets['juicio'] == 1]
for i in range(50):
    print('-' * 60)
    get_summary_of_tweet(juicio_tweets.iloc[i])
    print('-' * 60, '\n')
```

```python
print('Tweets with ex_pareja {}'.format(original_tweets['ex_pareja'].sum()))
ex_pareja_tweets = original_tweets[(original_tweets['ex_pareja'] == 1) & (original_tweets['lo_cuento_yo_porque'] == 1) ]
for i in range(50):
    print('-' * 60)
    get_summary_of_tweet(ex_pareja_tweets.iloc[i])
    print('-' * 60, '\n')
```
